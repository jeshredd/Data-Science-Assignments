```python
# 2. Encoding Techniques:
cat_cols=df_cleaned.select_dtypes(include='object').columns
cat_cols
```

```python
print(df_cleaned['workclass'].value_counts())
print(df_cleaned['education'].value_counts())
print(df_cleaned['marital_status'].value_counts())
print(df_cleaned['occupation'].value_counts())
print(df_cleaned['relationship'].value_counts())
print(df_cleaned['race'].value_counts())
print(df_cleaned['sex'].value_counts())
print(df_cleaned['native_country'].value_counts())
print(df_cleaned['income'].value_counts())
```

```python
# only sex and income columns have classes less than 5 thus encodig them with one hot encoder
one=OneHotEncoder()
onehot = pd.get_dummies(df_cleaned[['sex','income']],dtype=int)
onehot
```

```python
# label encoder
label=LabelEncoder()
df_cleaned['education'] = label.fit_transform(df_cleaned['education'])
df_cleaned['workclass'] = label.fit_transform(df_cleaned['workclass'])
df_cleaned['marital_status'] = label.fit_transform(df_cleaned['marital_status'])
df_cleaned['occupation'] = label.fit_transform(df_cleaned['occupation'])
df_cleaned['relationship'] = label.fit_transform(df_cleaned['relationship'])
df_cleaned['race'] = label.fit_transform(df_cleaned['race'])
df_cleaned['native_country'] = label.fit_transform(df_cleaned['native_country'])
df_cleaned.drop(df_cleaned[['sex','income']],axis=1,inplace=True)
df_cleaned=df_cleaned.join(onehot,how='right')
df_cleaned
```

```python
# label encoder:
#pros
#Simple and memory-efficient: Only one column per feature.
#Works well with tree-based models (e.g., Decision Trees, Random Forests, XGBoost)
#Cons:
#Imposes ordinal relationships between categories, even if they are nominal (unordered).
#Can bias models that assume numerical relationships, like Linear Regression or KNN.

#one hot coder:
#pros
#No ordinal relationships imposed, so its safe for nominal data
#Works well with linear models and distance-based algorithms like KNN.
#Cons:
#Increases dimensionality, especially with high-cardinality features (lots of unique values).
#which can increase memory usage and computation time.
```

```python
# 3. Feature Engineering:
# this feature captures the overall financial benefit or loss from capital activities.
#A positive value may indicate higher income and financial knowledge.
df_cleaned['capital_net_gain']=df_cleaned['capital_gain']-df_cleaned['capital_loss']
df_cleaned
```

```python
#This feature measures the relative educational attainment for a person's age.
#A higher value might indicate that a person has achieved more education at a younger age, which could be correlated with higher income potential.
df_cleaned['education_per_year']=df_cleaned['education']/df_cleaned['age']
df_cleaned
```

```python
# Applying log(1 + x) transformation to capital_gain
# capital_gain has the highest skewness to reduce the skewness we are applying this log transformation
df_cleaned['capital_gain_log'] = np.log1p(df_cleaned['capital_gain'])
df_cleaned
```

```python
# 4. Feature Selection:
forest=IsolationForest(random_state=100,contamination=0.1)
forest.fit(df_cleaned)
```

```python
outliers=pd.DataFrame(forest.predict(df_cleaned))
outliers.columns = ['outliers']
df_cleaned=df_cleaned.join(outliers,how='right')
df_cleaned
```

```python
# if predicted value is -1 the it is an outlier , if predicted value is 1 then it is an normal data point, thus only displaying the outliers
df_cleaned[df_cleaned['outliers']==-1]
```

```python
# removing the outliers
df_cleaned = df_cleaned[df_cleaned['outliers'] == 1].copy()
df_cleaned
```

```python
# outliers can Mislead Performance Metrics ie Outliers can inflate metrics like RMSE, giving a false impression of model performance.
#They can also reduce the accuracy of classification models, especially when class imbalance exists.
# Models may try to fit outliers, leading to overfitting.This reduces the modelâ€™s ability to generalize to new, unseen data.
```

```python
import ppscore as pps
```

```python
x=df.drop('income',axis=1) # dividing dependent and independent variables
y=df['income']
```

```python
pps_matrix = pps.matrix(df)

# Filtering for only rows where 'y' (target) is 'income' 
pps_income = pps_matrix[pps_matrix['y'] == 'income']

# Sorting by PPS score in descending order
pps_income_sorted = pps_income.sort_values(by='ppscore', ascending=False)

# Displaying the result
print(pps_income_sorted[['x', 'y', 'ppscore']])
```

```python
# the above analysis displays the top features that are strong predictors of the target variables based on their pps scores.
# PPS gives a better picture of predictive relationships than correlation, especially when dealing with categorical or non-linear variables.
# Features like capital_gain and education are shown as strong pedictors of income from both correlation and PPS.
```
