```python
# 1.EDA
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
df=pd.read_csv('Wine.csv')#importing the dataset
df.head()
```

```python
# visualization 
# creating histograms
for col in df.columns:
    df_cleaned = df.copy()
    df_cleaned[col].hist(bins=25)
    plt.ylabel('Count')
    plt.title(col)
    plt.show()
```

```python
# boxplots
cols=['Alcohol','Malic','Ash','Alcalinity','Magnesium','Phenols','Flavanoids','Nonflavanoids','Proanthocyanins',	'Color','Hue','Dilution','Proline']
for col in cols:
    series = df_cleaned[col]
  # finding outliers by iqr
    Q1 = series.quantile(0.25) # determining q1
    Q3 = series.quantile(0.75)  # determining q3
    IQR = Q3 - Q1
    lower_extreme = Q1 - 1.5 * IQR  #deterining lower extreme
    upper_extreme = Q3 + 1.5 * IQR   #deterining upper extreme
    outliers = series[(series < lower_extreme) | (series > upper_extreme)] # determining outliers

    plt.figure(figsize=(6, 4))
    df_cleaned.boxplot(column=col)
    plt.title(f"Boxplot of {col}")
    plt.xlabel(col)

    
    plt.tight_layout()
    plt.show()
    print(f"Outliers detected: {len(outliers)}") #printing no of outliers
    print(f"Lower extreme: {lower_extreme}")
    print(f"Upper extreme: {upper_extreme}")
```

```python
# generating correlation heat maps to explore relationships between pairs of variables.
# Computing the correlation matrix
corr_matrix = df_cleaned.corr()

# Setting  the matplotlib figure
plt.figure(figsize=(12, 10))

# plotting the heatmap with seaborn
sns.heatmap(corr_matrix, annot=True, fmt=".1f", cmap='coolwarm', square=True, linewidths=0.5)

plt.title("Correlation Heatmap ", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# standardizing the data
standard_scaler = StandardScaler()
std_data = standard_scaler.fit_transform(df_cleaned)
std_data.shape
```

```python
# Dimensionality Reduction with PCA:
pca = PCA()
pca_values = pca.fit_transform(std_data)
```

```python
pca_values
```

```python
# calculating explained variance
var = pca.explained_variance_ratio_
var
```

```python
# cumulative variance
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
# determining optimal no of principal components from the above analysis we are taking the first three . 
pca = PCA(n_components=3)
pca_std = pca.fit_transform(std_data)

# Converting to data frame
pca_std_data = pd.DataFrame(data = pca_std, columns = ['PC1', 'PC2','PC3'])
print(pca_std_data.shape)
pca_std_data.head() # this is the pca transformed dataset.
```

```python
# pca plot (pca-1,pca-2)
plt.figure(figsize=(9,6))

# Scatterplot
plt.scatter(pca_std_data.iloc[:,0], pca_std_data.iloc[:,1], s=40)

# Aesthetics
plt.title('PCA plot ',fontweight='bold',fontsize=18)
plt.xlabel('PC1',fontweight='bold', fontsize=16)
plt.ylabel('PC2',fontweight='bold', fontsize=16)
plt.show()
```

```python
#3: Clustering with Original Data:
# k-means clustering
# from elbow graph we can choose k value as 2,3,4,5 we will set k value after checking silhouette score
import warnings
warnings.filterwarnings('ignore')
wcss=[]
for i in range (1,9):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(std_data)
    wcss.append(kmeans.inertia_)
    
# Plotting K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
fig, ax = plt.subplots(figsize=(12,5))
plt.plot(range(1,9), wcss,color = 'black')
plt.scatter(range(1,9), wcss,color='red')
plt.title('Elbow Graph ')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#calculating Silhouette score to decide no of clusters ie k value
from sklearn.metrics import silhouette_score
clust = [2,3,4,5,6,7,8,9]
for n_clusters in clust:
    cluster1 = KMeans(n_clusters=n_clusters, random_state=0)
    cluster_labels1 = cluster1.fit_predict(std_data)
    score1= silhouette_score(std_data, cluster_labels1)
    print("For n_clusters =", n_clusters,"The silhouette_score is :", score1)
```

```python
# from the above analysis we can choose k value as 2,3,4,5 but we are choosing the no of clusters as 3 because it has the best silhouette score, thus deploying the model with it.
modelk = KMeans(n_clusters=3)
y_predk = modelk.fit_predict(std_data)
y_predk.shape
```

```python
y_predk
```

```python
# adding k cluster results to the original dataset
df_cleaned['kcluster_id']=modelk.labels_
df_cleaned
```

```python
#Cluster Analysis and Interpretation:
kcluster_summary = df_cleaned.groupby('kcluster_id').mean()
kcluster_summary
```

```python
# This model has created 3 clusters:
#Cluster 0
#Lowest Alcohol  but highest Malic Acid .
#Lowest Flavanoids and lowest Proline 
#Highest Color Intensity but lowest Hue 
#Lowest Hue suggests  less aging.
#This cluster represents cheaper or lower-end wines.

#Cluster 2 :
#Moderate Alcohol  and Moderate Proline 
#High Hue  which may suggest older or better-aged wine.
#Medium levels of Flavanoids  and Phenols 
#Balanced Color Intensity 
#This cluster seems to represent medium-quality wines that are not too strong or too weak .

#Cluster 1 
#Highest Alcohol  and very high Proline , high-quality wine.
#High Flavanoids  and Phenols , strong flavor profile
#High Color Intensity  and high Hue 
#Lowest Non-Flavanoids  , lower bitterness
#This cluster represents premium wines .

```

```python
# Visualization:
df_cleaned['kcluster_id'] = y_predk

# Selecting two features for visualization
x_feature = 'Type'
y_feature = 'Alcohol'

# Plotting the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_cleaned,
    x=x_feature,
    y=y_feature,
    hue='kcluster_id',
    palette='Set2',
    s=80,
    edgecolor='black'
)

plt.title(f"Cluster Visualization: {x_feature} vs {y_feature}", fontsize=16)
plt.xlabel(x_feature)
plt.ylabel(y_feature)
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
# 4. Clustering with PCA Data:
# k-means clustering
# from elbow graph we can choose k value as 2,3,4,5 we will set k value after checking silhouette score
import warnings
warnings.filterwarnings('ignore')
wcss=[]
for i in range (1,9):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(pca_std_data)
    wcss.append(kmeans.inertia_)
    
# Plotting K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
fig, ax = plt.subplots(figsize=(12,5))
plt.plot(range(1,9), wcss,color = 'black')
plt.scatter(range(1,9), wcss,color='red')
plt.title('Elbow Graph ')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#calculating Silhouette score to decide no of clusters ie k value
from sklearn.metrics import silhouette_score
clust = [2,3,4,5,6,7,8,9]
for n_clusters in clust:
    cluster2 = KMeans(n_clusters=n_clusters,random_state=0)
    cluster_labels2 = cluster2.fit_predict(pca_std_data)
    score2= silhouette_score(pca_std_data, cluster_labels2)
    print("For n_clusters =", n_clusters,"The silhouette_score is :", score2)
```

```python
# from the above analysis we can choose k value as 2,3,4,5 but we are choosing the no of clusters as 3 because it has the best silhouette score, thus deploying the model with it.
modelkp = KMeans(n_clusters=3)
y_predkp = modelkp.fit_predict(pca_std_data)
y_predkp
```

```python
# adding k cluster(pca) results to the original dataset
df_cleaned['kcluster_id_pca']=modelkp.labels_
df_cleaned
```

```python
#Cluster Analysis and Interpretation:
kclusterp_summary = df_cleaned.groupby('kcluster_id_pca').mean()
kclusterp_summary
```

```python
# This model has created 3 clusters:
#Cluster 2
#Highest Malic Acid 
#Lowest Flavanoids  and lowest Hue , less aged 
#Highest Color Intensity  
#Moderate Alcohol  and Proline 
#This cluster represents lowerend or inexpensive wines.

#Cluster  0
#Lowest Alcohol and Color Intensity 
#Moderate Phenols  and Flavanoids 
#High Hue  ,some aging
#Lowest Proline 
#This cluster represents medium-premium wines.

#Cluster 1
#Highest Alcohol  and highest Proline , suggests premium or aged wines
#Highest Flavanoids  and Phenols ,rich in taste and aroma
#High Color Intensity and Hue , well-aged
#This cluster contains high-quality or premium wines.

```

```python
# Visualization:
df_cleaned['kcluster_id_pca'] = y_predkp

# Selecting two features for visualization
x_feature = 'Type'
y_feature = 'Alcohol'

# Plotting the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_cleaned,
    x=x_feature,
    y=y_feature,
    hue='kcluster_id_pca',
    palette='Set2',
    s=80,
    edgecolor='black'
)

plt.title(f"Cluster Visualization: {x_feature} vs {y_feature}", fontsize=16)
plt.xlabel(x_feature)
plt.ylabel(y_feature)
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
# 5: Comparison and Analysis:
# In the Original Dataset cluster-0 represents lower wines , cluster-1 represents premium wines , cluster-2 represents moderately premium wines.	
# In the pca transformed dataset cluster-0 represents moderately premium wines, cluster-1 represent premium wines, cluster-2 lower end wines.
#All three clusters are still distinguishable in both approaches.
#Clusters maintain similar profiles in terms of wine quality tiers: low-end, medium, premium in both approaches.
#Cluster label numbers differ in both approaches — Cluster 0 in one model might correspond to Cluster 2 in another.
#The boundaries between medium- and low-quality wines are varied in PCA-transformed clustering when compared with original dataset(clustering).
# Raw data may include noise where as PCA filters out less useful info.
# the pca approach has high silhouette score compared with the other approach.
```

```python
#6. Conclusion and Insights
# Three different clusters were identified from the wine dataset using KMeans clustering, both on the original and PCA-transformed data.
# pca Reduces dimensionality, helping eliminate noise and redundancy
# k-means Helps uncover hidden patterns or groupings in unlabeled datasets.
# when you have high dimensional data  Use PCA first to reduce noise and improve clustering performance.
# in case of high dimensions for visualizating ,Use PCA to reduce to 2 or 3 dimensions for plotting.
```
