```python
# Data Preprocessing:
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
df = pd.read_excel('Online retail.xlsx',header=None)
df
```

```python
# converting the the data to appropriate format for Association Rule Mining:
df=pd.get_dummies(df,dtype='int')
df
```

```python
df.isnull().sum() # to check for missing values
```

```python
# Association Rule Mining:
# Apriori Algorithm
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
sorted_rules=rules.sort_values('lift',ascending = False)
sorted_rules
```

```python
# Displaying the first row as an association rule because it has the best support , confidence , lift values
row = sorted_rules.iloc[1]

antecedents =list(row['antecedents'])
consequents = list(row['consequents'])
confidence = row['confidence']
support = row['support']
lift = row['lift']

print(f"Rule: If a customer buys {{{antecedents}}}, they are likely to also buy {{{consequents}}}")
print(f"- Support: {support:.4f}")
print(f"- Confidence: {confidence:.2%}")
print(f"- Lift: {lift:.2f}")
```

```python
#Analysis and Interpretation:
# If a customer buys mineral water, they are likely to also buy spaghetti
# Customers buying ground beef are also buying spaghetti.
# Customers buying shrimp often add frozen vegetables.
# Many associations suggest meal-focused purchases:
#1.Shrimp + frozen vegetables.
#2Ground beef + spaghetti.
#3Mineral water + basic staples.
# insights for store 
#1. Place items like mineral water near spaghetti, or shrimp near frozen veggies.
#2. Offer meal bundles (e.g., “Dinner for 2” with shrimp, frozen veggies, and rice).

```
